#### Making Decimal Numbers Using Math

Let's make 2 into binary using math. Let's divide 2 by 2 until the result is 0.

$$ 2 \div 2 = 1 \text{ remainder } 0$$

We still haven't reached 0 yet. Divide by 2 one more time.

$$ 1 \div 2 = 0 \text{ remainder } 1$$

Then we can see that the first remainder **1** and second remainder **0** make up $10_2$ or 2.

That's all fine and dandy but how can we get this in code?

````python
n = 2 # number we are trying to convert
base = 1 # where we are in the number we are creating
result = 0 # where we copy the number into


result = result + base * (n % 2)
base = base * 10
n = n // 2
print(f"Number so far: {result}")
result = result + base * (n % 2)
base = base * 10
n = n // 2
print(f"Number so far: {result}")
````

Things to pay attention to are the modulus operator (`%`) in the `result`. It allows us to extract the **remainder** of the calculation. It extracts the current digit to add to the `result`.

Another thing to pay attention is the use of `//`. In python, the `/` in this problem would've given us trouble becuase it brings along the remainder. The `//` allows us to only get the integer part of the **quotient**

### TRIVIA!: Branchless programming



**Do not try this at home, but do at a friends house**

In programming there are things called `conditional statements` and we&rsquo;ll get to those later.

However, just for fun, you can mimic conditional statements with just math.

In this example, we have a number $321$ and a variable named $x$

if $x$ is **negative** then it evaluates to -321

if $x$ is **positive** then it evaluates to 321

$$ \begin{cases}
      -321 & x < 0 \\
      321 & x\geq 0
   \end{cases}
$$

where $x < 0 = 0$ and $x >= 0 = 1$



````python
x = 123
# 0 * -321 + 1 * 321
(x < 0) * -321 + (x >= 0) * 321
````

This works because in Python, the expression `(x < 0)` returns a $1$ or $0$

Now, lets change it such that $x < 0 = 1$ and $x >= 0 = 0$



````python
x = -123
# 1 * -321 + 0 * 321
(x < 0) * -321 + (x >= 0) * 321
````
