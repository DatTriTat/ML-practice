```python
# what types are all of the columns? and how many samples do you have?
data_types = salary_data.dtypes
num_samples = salary_data.shape[0]
print("Data Types:")
print(data_types)
print(f"\nTotal number of samples: {salary_data.shape[0]}")
```

```python
# remove any rows with an empty or null cell. How many did you remove?
initial_count = salary_data.shape[0]
salary_data.dropna(inplace=True)
rows_removed = initial_count - salary_data.shape[0]
print("Number of rows removed:", rows_removed)
```

```python
# drop all of the columns except 'Age', 'Gender','Job Title'
data_reduced = salary_data[['Age', 'Gender', 'Job Title', 'Salary']]
data_reduced.head()
```

```python
# convert the categorial data into numbers using this function
from sklearn.preprocessing import LabelEncoder
encoded_data = data_reduced.copy()

# Initialize LabelEncoder
le = LabelEncoder()

# Encode the 'Gender' column and 'Job Title' column
encoded_data['Gender'] = le.fit_transform(encoded_data['Gender'])
encoded_data['Job Title'] = le.fit_transform(encoded_data['Job Title'])

print(encoded_data.head())
```

```python
# split the data into training and testing dataset. use 80/20 split
from sklearn.model_selection import train_test_split
train_data, test_data = train_test_split(encoded_data, test_size=0.20, random_state=42)  # random_state for reproducibility
```

```python
# train the linear regression model from sklearn
from sklearn.linear_model import LinearRegression

# Prepare features and target
x_train = train_data[['Age', 'Gender', 'Job Title']]
y_train = train_data['Salary']
x_test = test_data[['Age', 'Gender', 'Job Title']]
y_test = test_data['Salary']

# Initialize the Linear Regression model
model = LinearRegression()

# Train the model
model.fit(x_train, y_train)
```

```python
from sklearn.metrics import mean_absolute_error, mean_absolute_percentage_error, r2_score
y_pred = model.predict(x_test)
mae = mean_absolute_error(y_test, y_pred)
mape = mean_absolute_percentage_error(y_test, y_pred) * 100

print(f"\nMean Absolute Error (MAE): {mae}")
print(f"Mean Absolute Percentage Error (MAPE): {mape}%")
```

```python
# what is the percentage of your accurracy?
r2 = r2_score(y_test, y_pred)
accuracy_percentage = r2 * 100
print(f"Model Accuracy: {accuracy_percentage:.2f}%")
```
