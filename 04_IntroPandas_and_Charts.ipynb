## Data Frames
They are objects that hold multiple series. To create a dataframe we need two data types:

1.   Using one or more `Series`.
2.   A dictionary.

If the `Series` don't match in length, missing values are automatically filled with special [NaN](http://pandas.pydata.org/pandas-docs/stable/missing_data.html) values.  

```python
city_names = pd.Series(['San Francisco', 'San Jose', 'Sacramento'])
population = pd.Series([852469, 1015785, 485199])
```

```python
print(city_names)
```

```python
print(population)
```

```python
# Dictionary is used to give each of the series a title
dict_df = { 'City name': city_names, 'Population': population }
df = pd.DataFrame(dict_df)

print('Dictionary:\n', dict_df)
print('\n Data Frame:\n', df)
```

```python
# Same code as above but removes the need for dict_df variable
cities = pd.DataFrame({ 'City name': city_names, 'Population': population })
print('\n Data Frame:\n', cities)
```

```python
cities.head()  # allows you to see the first few lines of the dataframe
```

```python
# create dataframe with City name, Population and Weather columns by adding a
# Weather column to the data frame we already created above

weather = pd.Series([77,96,109])
cities['Weather'] = weather
```

```python
cities.head()
```

For more complex single-column transformations, you can use `Series.apply`.
`Series.apply` accepts as an argument a [lambda function](https://docs.python.org/2/tutorial/controlflow.html#lambda-expressions), which is applied to each value.  

A lambda function is a one-line function with no name.

The example below creates a new `Series` that indicates whether `population` is over one million.  The way the below code works is that all the data in "population" is passed into the variable "val" and then the lamda function is applying the condition on "val" which returns True for rows for which the condition holds and False for rows for which the condition does now hold.

You can give any name to the variable which captures the data from population, e.g., instead of "val", you can call it "pop", so long as the variable name in the function body (after the colon), is the same as that specified in the function parameter (before the colon).

```python
# output population
population
```

```python
population.apply(lambda val: val > 1000000)
```

## Example 1: Access to healthcare

```python
df_health = pd.read_csv(housing+"HealthDataWeek30.csv") # note how we use string concatenation to combine
                                                        # the directory path with the file name

# inspect the first few lines of the data frame
df_health.head(5)
```

```python
# Sometimes there are white spaces in the original data file that may interfere with some of our processing,
# so we need to strip them first

# Strip the white spaces from the values in the Characteristics column and store the new versions in the dataframe
# (i.e., overwrite the old column with the new column)
df_health["Characteristics"] = df_health["Characteristics"].str.strip()
```

```python
# Same as above but applied to the Type column
df_health["Type"] = df_health["Type"].str.strip()
```

```python
# Due to the presence of commas in the numbers in the dataset (e.g., 1,000,000), they were initially loaded by Python as text strings.
# In order to apply mathematical operations to them, we need to remove the commas and convert the strings to integers

df_health["Delayed-Yes"] =  df_health["Delayed-Yes"].str.replace(",","").astype(int) # Change the "Delayed-Yes" column datatype to integer
df_health["Delayed-No"] = df_health["Delayed-No"].str.replace(",","").astype(int)    # Change the "Delayed-No" column datatype to integer

df_health.head(5)
```

```python
# Create a new dataframe out of the rows in the dataset which have a value of "Hispanic origin and Race" in the "Characteristics" column
# These are the rows that show the data categorized by race.
df_health_race = df_health[df_health["Characteristics"]=="Hispanic origin and Race"]

df_health_race.head(5) # Displays the top 5 rows in a dataframe
```

```python
# Create a pie chart
# The first argument to the function is the dataframe from which data will be extracted for display in the figure
# The "values" arument specifies the data frame column whose data is to be displayed
# The "names" argument specifies the data frame column which contains the label for the particular data
# (i.e., which slice of the pie the particular data point should be added to, e.g., in this case which race)

fig = px.pie(df_health_race, values='Delayed-Yes', names='Type', title='Delayed Healthcare by Race',color_discrete_sequence=px.colors.sequential.RdBu) # Plotting a pie chart

# Display the chart
fig.show()
```

## Example 2: Incarceration trends

```python
# Read the .csv file into a Dataframe called df_trends
df_trends = pd.read_csv(incarceration+"trends_incarceration.csv")

# Display top 5 rows
df_trends.head(5)
```

```python
# We want to re-organize the data so that all the rows for a particular crime (e.g., Possession) are summed
# for each year in order to display the total number of crimes for each type per year
# To do that, we create a new Data Frame called df_trends_grp and place in it the re-organized Data Frame

# In the code, we refer to the column names on which the operations are to be performed, e.g., "Crime Type", "Year", "Count"
# The periods in the notation below indicate successive operations applied to the data, e.g.,
# first we apply a groupby operation, then we apply a sum operation, and finally we apply a reset_index() operation
# The reset_index() operation creates a sequential numerical index for the Data Frame, e.g., 0, 1, 2...


df_trends_grp = df_trends.groupby(["Crime Type","Year"])["Count"].sum().reset_index()

# take a look at the new frame we created
df_trends_grp.head()
```

```python
# Plotting a stacked bar chart where the stacking is by "Crime Type",
# the x axis is the "Year" column, and the y axis is the "Count" column

fig = px.bar(df_trends_grp, x="Year", y="Count", color="Crime Type", title="Total crimes over years")
fig.show() # Display the fig variable
```

##Example 3: Eviction likelihood by state

```python
# Read .csv file into a Dataframe called df_surveydata
df_surveydata = pd.read_csv(housing+"HouseholdSurveyWeek30.csv") # note how we use string concatenation to combine
                                                          # the directory path with the file name

# Display top 5 rows of the dataframe
df_surveydata.head(5)
```

```python
# Read .csv file into a Dataframe called df_coords
df_coords = pd.read_csv(housing+"statelatlong.csv")

# inspect the contents of the Datat Frame
df_coords.head(5)
```

```python
# df_coords contains the Latitude & Longitude of the varioius states, while
# df_surveydata contains the rest of the data we are interested in
# We want to add the coordinates to df_surveydata so that we have coordinates
# for each state

# To do that, we will merge the df_coords data frame into the df_surveydata data frame,
# creating a df_merged_data data frame

# The columns are merging based on are "Area" in df_surveydata and "Statename" in df_coords
# (they both have the same set of values, which are the names of all the states)
# This means that the row which contains Alabama in df_surveydata is concatenated with
# the row which contains Alabama in df_coords

df_merged_data = df_surveydata.merge(df_coords, left_on="Area", right_on="Statename")

df_merged_data.head(5)
```

```python
# Dropping (removing) the"State" and "Statename" columns from the df_merged_data Data Frame
df_merged_data.drop(columns=["State","Statename"], inplace=True)

df_merged_data.head(5)
```

```python
fig = px.scatter_geo(df_merged_data,
                    lat=df_merged_data["Latitude"],
                    lon=df_merged_data["Longitude"],
                    hover_name="Area",
                    size="Likelihood of Eviction or Foreclosure Percent",
                    projection="albers usa",
                    title = "Likelihood of Eviction of household or Foreclosure Percent in USA") # Plotting the bubble map with the bubble siz being the Foreclosure Percent
fig.show() # Display the map
```
